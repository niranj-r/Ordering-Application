```python
import tkinter as tk
from tkinter import messagebox, simpledialog

# Initialize the medicine data
mk = {'Ibuprofen': 1, 'Advil': 1, 'Motrin': 1, 'Acetaminophen': 2, 'Tylenol': 2, 'Paracetamol': 2, 'Aspirin': 3,
      'Lisinopril': 4, 'Levothyroxine': 5, 'Synthroid': 5, 'Metformin': 6, 'Atorvastatin': 7, 'Lipitor': 7,
      'Omeprazole': 8, 'Prilosec': 8, 'Amlodipine': 9, 'Ventolin': 10} 

ml = {'Ibuprofen': 10, 'Advil': 10, 'Motrin': 10, 'Acetaminophen': 5, 'Tylenol': 5, 'Paracetamol': 5, 'Aspirin': 15,
      'Lisinopril': 8, 'Levothyroxine': 25, 'Synthroid': 25, 'Metformin': 50, 'Atorvastatin': 100, 'Lipitor': 100,
      'Omeprazole': 65, 'Prilosec': 65, 'Amlodipine': 75, 'Ventolin': 90}

mi = {1: 'A nonsteroidal anti-inflammatory drug (NSAID) commonly used to relieve pain, reduce inflammation, and lower fever.',
      2: 'A pain reliever and fever reducer widely used for headaches, muscle aches, arthritis, and reducing fever.',
      3: "NSAID used to relieve pain, reduce inflammation, and prevent blood clotting. It's commonly used for headaches, minor aches, and as a preventive measure for heart attacks and strokes.",
      4: "An ACE inhibitor used to treat high blood pressure (hypertension), heart failure, and to improve survival after heart attacks.",
      5: "A synthetic form of the thyroid hormone thyroxine, used to treat hypothyroidism (underactive thyroid).",
      6: "An oral medication used to treat type 2 diabetes by lowering blood sugar levels.",
      7: "A statin medication used to lower cholesterol levels and reduce the risk of cardiovascular disease.",
      8: "A proton pump inhibitor (PPI) used to reduce stomach acid production, treating conditions such as gastroesophageal reflux disease (GERD) and ulcers.",
      9: "A calcium channel blocker used to treat high blood pressure (hypertension) and chest pain (angina).",
      10: "A bronchodilator used to treat asthma and chronic obstructive pulmonary disease (COPD) by opening the airways in the lungs."}

# Main application class
class MedicineDeliveryApp:
    def __init__(self, root):
        self.root = root
        self.root.title("Medicine Delivery System")
        self.create_widgets()
        self.mcart = []
        self.qtcart = []
        self.nt = 0

    def create_widgets(self):
        self.label = tk.Label(self.root, text="Welcome to Medicine Delivery System!", font=("Helvetica", 16))
        self.label.pack(pady=20)

        self.order_button = tk.Button(self.root, text="Order Medicine", command=self.order_medicine, width=20)
        self.order_button.pack(pady=10)

        self.info_button = tk.Button(self.root, text="Know About Medicine", command=self.know_about_medicine, width=20)
        self.info_button.pack(pady=10)

        self.exit_button = tk.Button(self.root, text="Exit", command=self.root.quit, width=20)
        self.exit_button.pack(pady=10)

    def order_medicine(self):
        self.nt = 0
        self.mcart = []
        self.qtcart = []
        while True:
            m = simpledialog.askstring("Order Medicine", "Enter the medicine you would like to order:").title()
            if m in ml.keys():
                nq = simpledialog.askinteger("Order Medicine", "Enter the required quantity:")
                self.mcart.append(m)
                self.qtcart.append(nq)
                t = ml.get(m)
                ta = nq * t
                self.nt += ta + (0.1 * ta)
                messagebox.showinfo("Order Medicine", f'Your order is for {self.qtcart} {self.mcart}, the current total is: {self.nt}')
            else:
                messagebox.showwarning("Order Medicine", "Medicine not available!")
            
            a = simpledialog.askstring("Order Medicine", "Do you wish to continue? (yes/no)").lower()
            if a != "yes":
                break

        inp = simpledialog.askinteger("Order Medicine", "Do you wish to place the order? If yes, press 1")
        if inp == 1:
            messagebox.showinfo("Order Medicine", "Order placed successfully!")

    def know_about_medicine(self):
        while True:
            m = simpledialog.askstring("Know About Medicine", "Enter the medicine you would like to know about:").title()
            if m in ml.keys():
                infokey = mk.get(m)
                messagebox.showinfo(m, mi.get(infokey))
            else:
                messagebox.showwarning("Know About Medicine", "Medicine not available in database!")
            
            a = simpledialog.askstring("Know About Medicine", "Do you wish to continue? (yes/no)").lower()
            if a != "yes":
                break

# Main application entry point
if __name__ == "__main__":
    root = tk.Tk()
    app = MedicineDeliveryApp(root)
    root.mainloop()
```
