```python
import tkinter as tk
from tkinter import messagebox

class LoginApp:
    def __init__(self, root):
        self.root = root
        self.root.title("Login System")
        self.users = {'Niranj': '1703', 'Gowtham': '1010', 'test': '0'}  # Initial user data
        self.create_widgets()

    def create_widgets(self):
        self.main_frame = tk.Frame(self.root)
        self.main_frame.pack(pady=20)

        self.title_label = tk.Label(self.main_frame, text="Welcome to the Login System", font=("Helvetica", 16))
        self.title_label.grid(row=0, columnspan=2, pady=10)

        self.signup_button = tk.Button(self.main_frame, text="Sign Up", command=self.sign_up)
        self.signup_button.grid(row=1, column=0, padx=20, pady=10)

        self.login_button = tk.Button(self.main_frame, text="Log In", command=self.log_in)
        self.login_button.grid(row=1, column=1, padx=20, pady=10)

        self.exit_button = tk.Button(self.main_frame, text="Exit", command=self.root.quit)
        self.exit_button.grid(row=2, columnspan=2, pady=10)

    def sign_up(self):
        self.clear_frame()
        self.title_label.config(text="Sign Up")

        self.username_label = tk.Label(self.main_frame, text="Username:")
        self.username_label.grid(row=1, column=0, padx=10, pady=10)
        self.username_entry = tk.Entry(self.main_frame)
        self.username_entry.grid(row=1, column=1, padx=10, pady=10)

        self.password_label = tk.Label(self.main_frame, text="Password:")
        self.password_label.grid(row=2, column=0, padx=10, pady=10)
        self.password_entry = tk.Entry(self.main_frame, show="*")
        self.password_entry.grid(row=2, column=1, padx=10, pady=10)

        self.signup_confirm_button = tk.Button(self.main_frame, text="Sign Up", command=self.perform_signup)
        self.signup_confirm_button.grid(row=3, columnspan=2, pady=10)

        self.back_button = tk.Button(self.main_frame, text="Back", command=self.go_back)
        self.back_button.grid(row=4, columnspan=2, pady=10)

    def perform_signup(self):
        username = self.username_entry.get()
        password = self.password_entry.get()

        if username in self.users:
            messagebox.showerror("Error", "Username already exists. Please choose a different one.")
        elif username == "" or password == "":
            messagebox.showerror("Error", "Username and password cannot be empty.")
        else:
            self.users[username] = password
            messagebox.showinfo("Success", "Sign up successful!")
            self.log_in()

    def log_in(self):
        self.clear_frame()
        self.title_label.config(text="Log In")

        self.username_label = tk.Label(self.main_frame, text="Username:")
        self.username_label.grid(row=1, column=0, padx=10, pady=10)
        self.username_entry = tk.Entry(self.main_frame)
        self.username_entry.grid(row=1, column=1, padx=10, pady=10)

        self.password_label = tk.Label(self.main_frame, text="Password:")
        self.password_label.grid(row=2, column=0, padx=10, pady=10)
        self.password_entry = tk.Entry(self.main_frame, show="*")
        self.password_entry.grid(row=2, column=1, padx=10, pady=10)

        self.login_confirm_button = tk.Button(self.main_frame, text="Log In", command=self.perform_login)
        self.login_confirm_button.grid(row=3, columnspan=2, pady=10)

        self.back_button = tk.Button(self.main_frame, text="Back", command=self.go_back)
        self.back_button.grid(row=4, columnspan=2, pady=10)

    def perform_login(self):
        username = self.username_entry.get()
        password = self.password_entry.get()

        if username in self.users and self.users[username] == password:
            messagebox.showinfo("Success", "Login successful!")
            self.selector()
        else:
            messagebox.showerror("Error", "Incorrect username or password.")

    def selector(self):
        self.clear_frame()
        self.title_label.config(text="Welcome!")

        self.welcome_label = tk.Label(self.main_frame, text="You have successfully logged in!", font=("Helvetica", 16))
        self.welcome_label.grid(row=0, columnspan=2, pady=20)

        self.exit_button = tk.Button(self.main_frame, text="Exit", command=self.root.quit)
        self.exit_button.grid(row=1, columnspan=2, pady=10)

    def go_back(self):
        self.clear_frame()
        self.create_widgets()

    def clear_frame(self):
        for widget in self.main_frame.winfo_children():
            widget.destroy()

if __name__ == "__main__":
    root = tk.Tk()
    app = LoginApp(root)
    root.mainloop()
```
