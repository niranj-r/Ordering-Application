```python
import tkinter as tk
from tkinter import messagebox

# Main Window
def main_menu():
    main_window = tk.Tk()
    main_window.title("Customer Support System")

    def open_support():
        main_window.destroy()
        supp()

    main_label = tk.Label(main_window, text="Welcome to the Customer Support System", font=('Helvetica', 16))
    main_label.pack(pady=20)

    support_button = tk.Button(main_window, text="Customer Support", command=open_support, width=20, height=2)
    support_button.pack(pady=10)

    main_window.mainloop()

# Support Menu
def supp():
    supp_window = tk.Tk()
    supp_window.title("Customer Support")

    def open_ai():
        supp_window.destroy()
        ai()

    def open_human():
        supp_window.destroy()
        hum()

    def go_back():
        supp_window.destroy()
        main_menu()

    label = tk.Label(supp_window, text="What do you want to start with?", font=('Helvetica', 16))
    label.pack(pady=20)

    ai_button = tk.Button(supp_window, text="Chat with CHERA (AI Bot)", command=open_ai, width=30, height=2)
    ai_button.pack(pady=10)

    human_button = tk.Button(supp_window, text="Chat with Human Assistant", command=open_human, width=30, height=2)
    human_button.pack(pady=10)

    back_button = tk.Button(supp_window, text="Go Back to Main Menu", command=go_back, width=30, height=2)
    back_button.pack(pady=10)

    supp_window.mainloop()

# Human Support
def hum():
    hum_window = tk.Tk()
    hum_window.title("Human Support Assistant")

    def go_back():
        hum_window.destroy()
        supp()

    label = tk.Label(hum_window, text="Welcome to NjAi's Human Support Assistant!", font=('Helvetica', 16))
    label.pack(pady=20)

    message_label = tk.Label(hum_window, text="Enter your message here:", font=('Helvetica', 12))
    message_label.pack(pady=10)

    message_entry = tk.Entry(hum_window, width=50)
    message_entry.pack(pady=10)

    def send_message():
        message = message_entry.get()
        if message:
            messagebox.showinfo("Message Sent", "Hey! Thanks for your message. We're routing you to a human representative who will address your needs promptly")
        else:
            messagebox.showerror("Error", "Please enter a message")

    send_button = tk.Button(hum_window, text="Send", command=send_message, width=20, height=2)
    send_button.pack(pady=10)

    back_button = tk.Button(hum_window, text="Go Back to Support Menu", command=go_back, width=30, height=2)
    back_button.pack(pady=10)

    hum_window.mainloop()

# AI Support
def ai():
    ai_window = tk.Tk()
    ai_window.title("CHERA - AI Bot")

    def go_back():
        ai_window.destroy()
        supp()

    label = tk.Label(ai_window, text="Welcome to CHERA powered by NjAi!", font=('Helvetica', 16))
    label.pack(pady=20)

    general_button = tk.Button(ai_window, text="General Queries", command=gq, width=30, height=2)
    general_button.pack(pady=10)

    availability_button = tk.Button(ai_window, text="Queries related to availability of a medicine", command=am, width=30, height=2)
    availability_button.pack(pady=10)

    order_button = tk.Button(ai_window, text="Queries related to your order", command=orde, width=30, height=2)
    order_button.pack(pady=10)

    payment_button = tk.Button(ai_window, text="Queries related to payment or refund", command=pr, width=30, height=2)
    payment_button.pack(pady=10)

    back_button = tk.Button(ai_window, text="Exit to Main Support menu", command=go_back, width=30, height=2)
    back_button.pack(pady=10)

    ai_window.mainloop()

# General Queries
def gq():
    gq_window = tk.Tk()
    gq_window.title("General Queries")

    def go_back():
        gq_window.destroy()
        ai()

    label = tk.Label(gq_window, text="General Queries", font=('Helvetica', 16))
    label.pack(pady=20)

    options = [
        "I received a discount code after placing my order. Can I apply it retroactively to receive the discount?",
        "I paid for my order, but I haven't received a receipt. Can you resend it to me?"
    ]

    var = tk.StringVar(value=options)

    listbox = tk.Listbox(gq_window, listvariable=var, height=6, selectmode=tk.SINGLE)
    listbox.pack(pady=10)

    def show_answer():
        selected = listbox.curselection()
        if selected:
            choice = selected[0]
            if choice == 0:
                messagebox.showinfo("Answer", "We appreciate your interest in using the discount code. Unfortunately, discount codes must be applied at the time of placing the order and cannot be applied retroactively. However, you can use the discount code for your next purchase. If you encounter any issues applying the code during checkout, please let us know, and we'll be happy to assist you.")
            elif choice == 1:
                messagebox.showinfo("Answer", "Certainly! We understand the importance of having a receipt for your order. Please provide us with your order details, including the order number and your registered email address. Our team will resend the receipt to your email promptly. If you have any other questions or concerns, feel free to let us know.")
        else:
            messagebox.showerror("Error", "Please select a query")

    select_button = tk.Button(gq_window, text="Select", command=show_answer, width=20, height=2)
    select_button.pack(pady=10)

    back_button = tk.Button(gq_window, text="Exit to Main Support menu", command=go_back, width=30, height=2)
    back_button.pack(pady=10)

    gq_window.mainloop()

# Queries related to availability of a medicine
def am():
    am_window = tk.Tk()
    am_window.title("Medicine Availability Queries")

    def go_back():
        am_window.destroy()
        ai()

    label = tk.Label(am_window, text="Queries related to availability of a medicine", font=('Helvetica', 16))
    label.pack(pady=20)

    queries = [
        "Is this medicine available?",
        "Do you have a particular medicine in stock?",
        "Are there any alternatives to this medicine?",
        "Can I get a refill of my prescription for a medicine?",
        "When will this medicine be restocked?"
    ]

    var = tk.StringVar(value=queries)

    listbox = tk.Listbox(am_window, listvariable=var, height=10, selectmode=tk.SINGLE)
    listbox.pack(pady=10)

    def show_answer():
        selected = listbox.curselection()
        if selected:
            query = listbox.get(selected[0])
            medicine = medicine_entry.get().title()

            mk = {'Ibuprofen': 1, 'Advil': 1, 'Motrin': 1, 'Acetaminophen': 2, 'Tylenol': 2, 'Paracetamol': 2, 'Aspirin': 3,
                  'Lisinopril': 4, 'Levothyroxine': 5, 'Synthroid': 5, 'Metformin': 6, 'Atorvastatin': 7, 'Lipitor': 7,
                  'Omeprazole': 8, 'Prilosec': 8, 'Amlodipine': 9, 'Ventolin': 10}
            ml = {'Ibuprofen': 10, 'Advil': 10, 'Motrin': 10, 'Acetaminophen': 5, 'Tylenol': 5, 'Paracetamol': 5, 'Aspirin': 15,
                  'Lisinopril': 8, 'Levothyroxine': 25, 'Synthroid': 25, 'Metformin': 50, 'Atorvastatin': 100, 'Lipitor': 100,
                  'Omeprazole': 65, 'Prilosec': 65, 'Amlodipine': 75, 'Ventolin': 90}
            mt = {1: 'Ibuprofen, Advil, Motrin', 2: 'Acetaminophen, Tylenol, Paracetamol', 3: 'Not Available', 4: 'Not Available',
                  5: 'Levothyroxine, Synthroid', 6: 'Not Available', 7: 'Atorvastatin, Lipitor', 8: 'Omeprazole, Prilosec', 9: 'Not Available',
                  10: 'Not Available'}

            if query == "Is this medicine available?":
                if medicine in mk:
                    messagebox.showinfo("Answer", f"Yes, {medicine} is currently available in our inventory.")
                else:
                    messagebox.showinfo("Answer", f"No, {medicine} is not currently available in our inventory. But you can check for alternatives!")

            elif query == "Do you have a particular medicine in stock?":
                if medicine in mk:
                    messagebox.showinfo("Answer", f"Yes, {medicine} is in stock and ready for delivery.")
                else:
                    messagebox.showinfo("Answer", f"No, {medicine} is not in stock. But you can check for alternatives!")

            elif query == "Are there any alternatives to this medicine?":
                if medicine in mk:
                    alternatives = mt[mk[medicine]]
                    messagebox.showinfo("Answer", f"Alternatives for {medicine} are: {alternatives}")
                else:
                    messagebox.showinfo("Answer", f"No alternatives available for {medicine}.")

            elif query == "Can I get a refill of my prescription for a medicine?":
                if medicine in mk:
                    messagebox.showinfo("Answer", f"Yes, we can arrange a refill of your prescription for {medicine}. Please provide your prescription details.")
                else:
                    messagebox.showinfo("Answer", f"We are extremely sorry, {medicine} is not currently available in our inventory. But you can check for alternatives!")

            elif query == "When will this medicine be restocked?":
                if medicine in mk:
                    messagebox.showinfo("Answer", f"Yes, {medicine} is currently available in our inventory.")
                else:
                    messagebox.showinfo("Answer", f"We expect to restock {medicine} within 7 business days. In the meantime, we can assist you with alternative options if needed.")

        else:
            messagebox.showerror("Error", "Please select a query and enter a medicine name")

    medicine_label = tk.Label(am_window, text="Medicine name:", font=('Helvetica', 12))
    medicine_label.pack(pady=10)

    medicine_entry = tk.Entry(am_window, width=50)
    medicine_entry.pack(pady=10)

    select_button = tk.Button(am_window, text="Select", command=show_answer, width=20, height=2)
    select_button.pack(pady=10)

    back_button = tk.Button(am_window, text="Exit to Main Support menu", command=go_back, width=30, height=2)
    back_button.pack(pady=10)

    am_window.mainloop()

# Queries related to your order
def orde():
    orde_window = tk.Tk()
    orde_window.title("Order Queries")

    def go_back():
        orde_window.destroy()
        ai()

    label = tk.Label(orde_window, text="Queries related to your order", font=('Helvetica', 16))
    label.pack(pady=20)

    options = [
        "What is the status of my order?",
        "When will my order be delivered?",
        "I've made the payment, but I haven't received any confirmation. Can you confirm if the payment was successful?",
        "I couldn't find a particular medicine in the app. Is it available?",
        "I need to make changes to my order / cancel my order. How can I do that?"
    ]

    var = tk.StringVar(value=options)

    listbox = tk.Listbox(orde_window, listvariable=var, height=10, selectmode=tk.SINGLE)
    listbox.pack(pady=10)

    def show_answer():
        selected = listbox.curselection()
        if selected:
            choice = selected[0]
            if choice == 0:
                messagebox.showinfo("Answer", "Your order is currently being processed and is expected to be dispatched within 1 day. You can track your order status in the app under 'My Orders'.")
            elif choice == 1:
                messagebox.showinfo("Answer", "Your order is scheduled for delivery on 11 May 2024 between 10am-10pm. Our delivery partner will notify you once the order is out for delivery.")
            elif choice == 2:
                messagebox.showinfo("Answer", "Thank you for your payment. We've received confirmation of your payment and your order is now being processed. You'll receive a confirmation email shortly.")
            elif choice == 3:
                messagebox.showinfo("Answer", "We apologize for the inconvenience. Please provide the name of the medicine, and we'll check its availability in our inventory. Alternatively, you can explore similar medicines as per your requirement.")
            elif choice == 4:
                messagebox.showinfo("Answer", "Sure, please provide your order details, and we'll assist you with the necessary modifications or cancellation. Please note that order modifications or cancellations are subject to our cancellation policy.")

        else:
            messagebox.showerror("Error", "Please select a query")

    select_button = tk.Button(orde_window, text="Select", command=show_answer, width=20, height=2)
    select_button.pack(pady=10)

    back_button = tk.Button(orde_window, text="Exit to Main Support menu", command=go_back, width=30, height=2)
    back_button.pack(pady=10)

    orde_window.mainloop()

# Queries related to payment or refund
def pr():
    pr_window = tk.Tk()
    pr_window.title("Payment or Refund Queries")

    def go_back():
        pr_window.destroy()
        ai()

    label = tk.Label(pr_window, text="Queries related to payment or refund", font=('Helvetica', 16))
    label.pack(pady=20)

    options = [
        "I was charged twice for my recent order. Can you please check and refund the extra amount?",
        "I received a notification that my payment failed, but I see the amount deducted from my account. When will I get my refund?",
        "I accidentally placed the wrong order and made the payment. Can I cancel the order and get a refund?",
        "I returned a product, but I haven't received the refund yet. Can you please update me on the refund status?",
        "I was overcharged for my order. How can I request a refund for the excess amount?"
    ]

    var = tk.StringVar(value=options)

    listbox = tk.Listbox(pr_window, listvariable=var, height=10, selectmode=tk.SINGLE)
    listbox.pack(pady=10)

    def show_answer():
        selected = listbox.curselection()
        if selected:
            choice = selected[0]
            if choice == 0:
                messagebox.showinfo("Answer", "We apologize for the inconvenience. Please provide us with your order details, including the order number and payment transaction IDs. Our team will investigate the issue promptly and initiate the refund for the duplicate charge.")
            elif choice == 1:
                messagebox.showinfo("Answer", "We regret the inconvenience caused. Payment failures can sometimes occur due to technical issues. Rest assured, if the payment was deducted from your account, it will be automatically refunded within 7 business days. Please check your bank statement for the refund transaction.")
            elif choice == 2:
                messagebox.showinfo("Answer", "We understand mistakes happen. If you wish to cancel your order, please reach out to our customer support team immediately with your order details. If the order hasn't been processed yet, we'll cancel it and initiate a refund to your original payment method.")
            elif choice == 3:
                messagebox.showinfo("Answer", "We appreciate your patience. Once we receive the returned product and verify its condition, we initiate the refund process. Refunds typically take 7 business days to reflect in your account, depending on your bank's processing time. You'll receive an email notification once the refund is processed.")
            elif choice == 4:
                messagebox.showinfo("Answer", "We apologize for the billing error. Please provide us with the relevant details, including your order number and the discrepancy in the amount charged. Our finance team will review the transaction and promptly process the refund for the excess amount to your original payment method.")

        else:
            messagebox.showerror("Error", "Please select a query")

    select_button = tk.Button(pr_window, text="Select", command=show_answer, width=20, height=2)
    select_button.pack(pady=10)

    back_button = tk.Button(pr_window, text="Exit to Main Support menu", command=go_back, width=30, height=2)
    back_button.pack(pady=10)

    pr_window.mainloop()

if __name__ == "__main__":
    main_menu()
```
