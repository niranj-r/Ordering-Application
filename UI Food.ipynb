```python
import tkinter as tk
from tkinter import messagebox

# Prices dictionary
prices = {"Pizza": 10, "Burger": 8, "Pasta": 12, "Sushi": 15, "Ramen": 11, "Tacos": 9, "Burritos": 10, "Quesadillas": 10}

# Sample restaurants dictionary
restaurants = {
    "Restaurant A": ["Pizza", "Burger"],
    "Restaurant B": ["Pasta", "Sushi"],
    "Restaurant C": ["Ramen", "Tacos", "Burritos", "Quesadillas"]
}

# Global food cart
foodcart = []

# Function to add items to cart
def add_to_cart(item):
    foodcart.append(item)
    cart_label.config(text="Cart: " + ', '.join(foodcart))

# Function to calculate bill
def calculate_bill():
    total = 0
    for item in foodcart:
        if item in prices:
            total += prices[item]
        else:
            messagebox.showerror("Error", "Something went wrong")
            return
    messagebox.showinfo("Total Bill", f"Total bill: ${total}")

# Function to display food items for the selected restaurant
def show_food_items():
    restaurant = restaurant_var.get()
    food_items = restaurants[restaurant]
    for widget in food_frame.winfo_children():
        widget.destroy()
    for item in food_items:
        button = tk.Button(food_frame, text=item, command=lambda item=item: add_to_cart(item))
        button.pack(side=tk.LEFT)

# Create the main window
root = tk.Tk()
root.title("Restaurant Menu")

# Create a frame for restaurant selection
restaurant_frame = tk.Frame(root)
restaurant_frame.pack(pady=10)

# Create and set the restaurant selection variable
restaurant_var = tk.StringVar(value="Restaurant A")
restaurant_label = tk.Label(restaurant_frame, text="Choose your restaurant:")
restaurant_label.pack(side=tk.LEFT)

# Create radio buttons for restaurant selection
for restaurant in restaurants.keys():
    radio = tk.Radiobutton(restaurant_frame, text=restaurant, variable=restaurant_var, value=restaurant, command=show_food_items)
    radio.pack(side=tk.LEFT)

# Create a frame for food items
food_frame = tk.Frame(root)
food_frame.pack(pady=10)

# Create a label to display the cart
cart_label = tk.Label(root, text="Cart: ")
cart_label.pack(pady=10)

# Create a button to calculate the bill
bill_button = tk.Button(root, text="Calculate Bill", command=calculate_bill)
bill_button.pack(pady=10)

# Initialize with the default restaurant's food items
show_food_items()

# Run the application
root.mainloop()
```
